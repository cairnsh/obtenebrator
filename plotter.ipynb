```python
import numpy as np, scipy as sp
import random
from importlib import reload
np.set_printoptions(precision=4, suppress=True)
```

```python
import matplotlib.pyplot as p
import obtenebrate
import testdata

RECOVERED = testdata.RECOVERED
ORIGINAL = testdata.ORIGINAL
```

```python
def plot(i):
    a = RECOVERED[i]
    b = ORIGINAL[i]
    x = []
    y = []
    def add(x, y, a):
        for j in range(len(a)):
            x.extend(a[j])
            y.extend([j] * len(a[j]))
    add(x, y, a)
    p.figure(figsize=(12, 12))
    p.scatter(x, y, color="red")
    for X in b:
        p.axvline(X, 0, 1)
    p.gca().axes.yaxis.set_visible(False)
    p.show()
```

```python
print("""
    The estimates were reconstructed from 4000000 samples of a mixture
    of 8 product distributions on 40 coordinates, with parameters chosen
    uniformly at random.
    
    In each graph, the blue lines are the 8 true parameters.
    Each horizontal line of red dots shows one group of estimates.
    """)

for i in range(40):
    print("Coordinate %d" % i)
    plot(i)
```
