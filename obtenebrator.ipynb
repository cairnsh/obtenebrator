We suppose that we have $r$ product distributions on $\{-1, 1\}^n$ with Bernoulli parameters $p^{(i)}_j$ where $i = 1, \ldots, r$ and $j = 1, \ldots, n$, and that they are combined with weights $w_1, \ldots, w_r$, $w_i > 0$, $w_1 + \cdots + w_r = 1$. We are allowed to draw samples from this distribution.

Let $x_1, \ldots, x_M \in \{-1, 1\}^n$ be samples and let $G(t_1, \ldots, t_n) = {1 \over M} \sum_{i=1}^M \prod (\frac12 + \frac12 x_i t_i)$ be the empirical distribution. The representation here is that if $x \in \{-1, 1\}^n$ then $G(x_1, \ldots, x_n)$ tells you the fraction of samples that are equal to $x$.

Let $h_i(t_1, \ldots, t_r) = \prod_{j=1}^n \left(\frac12 + (p^{(i)}_j - \frac12) t_j\right)$ be the real distributions and also we let $H = \sum_{i=1}^r w_i h_i$. We suppose that this is the minimal representation, i.e. that there is no choice of distributions $h'_1, \ldots, h'_{r'}$ with $H = \sum w'_i h'_i$ and $r' < r$. In particular this means that the span of $h_i$ has dimension exactly equal to $r$ because otherwise we could use a linear dependence to zero out one of the entries. We might also need an assumption that there's no choice of distributions with a weighted sum that's *approximately* $H$.

We are given a coordinate $j$ and parameters $n, \rho, r$. We are supposed to recover the set $\{p^i_j\}$. The algorithm works as follows.

- We pick a random selection of $\rho - 1$ coordinates from the remaining $n - 1$. (Suppose $j = 1$ and the remaining coordinates are $2, \ldots, \rho$.)
- Write $\vec G(x_1, \ldots, x_\rho) = G(x_1, \ldots, x_{\rho}, t_{\rho + 1}, \ldots, t_{n}) \in \mathbb C[t_{\rho + 1}, \ldots, t_{n}]/(t_{\rho + 1}^2, \ldots, t_n^2)$. (We hope that this has a low probability of decreasing the rank.)
- Do some kind of a projection step. (We have chosen to project $\prod_{i \in I} t_{i}$ to $\sum_{i} (2 \unicode{x1D7D9}_{i \in I} - 1) e_{i}$.)
- Let $q = 0, 0.0001, \ldots, 1$. Set
$$\mathcal{O}(q, x_2, \ldots, x_\rho) = q\vec G(0, x_2, \ldots, x_\rho) - (1-q)\vec G(1, x_2, \ldots, x_\rho).$$
This gives us a list of $2^{\rho - 1}$ vectors. We think of this as a matrix $\mathcal{O}(q)$ and use SVD to estimate the $r$-th singular value of it. Call that $s_r(q)$.
- Find the local minima of $s_r(q)$, i.e. the values of $q$ with $s_r(q - 0.0001) > s_r(q) < s_r(q + 0.0001)$. These numbers are the estimates.

Here $r$ is the rank known in advance (although we could definitely try to recover it from the SVDs) and $\rho$ is some small multiple of $\lg_2 r$.

- The columns of theq matrix $\mathcal{O}(q)$ are weighted sums of the vectors of Bernoulli parameters $2p^i_{\rho+1:n} - {\bf 1}$, plus a large amount of noise from the projections of the Boolean variables. My intuition is that there is a "noise floor" of singular values, and that we need enough samples so that the "real" singular values from the $r$-dimensional vectors won't get swamped by the noisy singular values.
